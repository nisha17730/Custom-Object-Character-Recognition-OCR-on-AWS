Custom OCR Project - Setup and Image Loading
------------------------------------------------
This script sets up the necessary paths, configures Tesseract, and loads an image for display.
It is a part of a larger project to extract text from lab reports using Tesseract OCR.

```python
# Import necessary libraries
import cv2  # OpenCV for image processing
import pytesseract  # Tesseract OCR for text extraction
import numpy as np  # NumPy for numerical operations
import pandas as pd  # Pandas for handling structured data
import os  # OS module for file handling
import matplotlib.pyplot as plt  # Matplotlib for image visualization
```

# Step 1: Define Paths

```python
# Path to the folder containing raw images
IMAGE_FOLDER = r"C:\Users\hp\project 10\raw_images"
# Path to the folder where output will be saved
OUTPUT_FOLDER = r"C:\Users\hp\project 10\output"

# Ensure the output folder exists
if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)
```

# Step 2: Configure Tesseract OCR

```python
# Define the path to the Tesseract executable (Update if necessary)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

# Step 3: Function to Load and Display an Image

```python
def load_and_display_image(image_path):
    '''
    This function loads an image from the given path, converts it to RGB format,
    and displays it using Matplotlib.
    '''
    image = cv2.imread(image_path)  # Load image
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB (Matplotlib uses RGB format)
    
    # Display the image
    plt.figure(figsize=(10, 5))
    plt.imshow(image)
    plt.axis("off")  # Hide axes
    plt.show()
    
    return image
```

# Step 4: Load and Display a Sample Image

```python
# Select a sample image from the raw images folder
sample_image_path = os.path.join(IMAGE_FOLDER, os.listdir(IMAGE_FOLDER)[0])
image = load_and_display_image(sample_image_path)

print("Setup complete. Image loaded successfully!")
```
